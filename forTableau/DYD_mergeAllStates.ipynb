# DYD Thermostat Data 

### Combine month CSV Files 
1. Read in files in folders for each state
2. Export as combined CSV to a folder not posted to GitHub due to file size



```python
# Dependencies
import pandas as pd
import os
import numpy as np
from pathlib import Path
from datetime import datetime
```

```python
# Create variable for files in directory
files = [f for f in os.listdir("data/") if f.endswith(".csv")]

# files
```

```python
# https://stackoverflow.com/questions/63886787/how-to-create-a-dataframe-from-multiple-csv-files
# Initialize an empty dataframe
AllStates = pd.DataFrame()

# Iterate through files and contents, then concatenate the data into the dataframe
for file in files:
    df = pd.read_csv("data/" + file)
    AllStates = pd.concat([AllStates, df])
    
AllStates
```

```python
AllStates.to_csv("../../data_large/AllStates.csv", header=True, index=False)
```
